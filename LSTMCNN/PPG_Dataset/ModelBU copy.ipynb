```python
# Define filtering functions
def lowpass_filter(data, cutoff=15, fs=2175, order=4):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low')
    return filtfilt(b, a, data)

def bandpass_filter(data, lowcut=0.5, highcut=8, fs=30, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, data)
```

```python
# Cell 2: Define data augmentation function
def apply_gaussian_noise(signal, noise_std=0.02):
    noise = np.random.normal(0, noise_std, signal.shape)
    return signal + noise
```

```python
def process_ppg_file(signal_path, label_path, target_sampling_rate=30, augment=False):
    signal_data = pd.read_csv(signal_path, header=None).values.flatten()
    label_data = pd.read_csv(label_path)
    
    filtered_signal = lowpass_filter(signal_data)
    original_sampling_rate = 2175
    num_samples = int(len(filtered_signal) * (target_sampling_rate / original_sampling_rate))
    downsampled_signal = resample(filtered_signal, num_samples)
    processed_signal = bandpass_filter(downsampled_signal, fs=target_sampling_rate)
    processed_signal = (processed_signal - np.mean(processed_signal)) / np.std(processed_signal)
    
    glucose_level = label_data['Glucose'].values[0]
    age = label_data['Age'].values[0]
    gender = 1 if label_data['Gender'].values[0] == 'Male' else 0
    
    augmented_signals = [processed_signal]
    
    if augment:
        for _ in range(3):  # Apply Gaussian noise 3 times per signal
            augmented_signals.append(apply_gaussian_noise(processed_signal))
    
    return [(sig, glucose_level, age, gender) for sig in augmented_signals]
```

```python
def process_all_ppg_files(root_folder, output_file, augment=False):
    signal_folder = os.path.join(root_folder, "RawData")
    label_folder = os.path.join(root_folder, "Labels")
    
    if not os.path.exists(signal_folder):
        raise FileNotFoundError(f"Error: Signal folder not found at {signal_folder}")
    if not os.path.exists(label_folder):
        raise FileNotFoundError(f"Error: Label folder not found at {label_folder}")
    
    processed_data = []
    
    for signal_file in os.listdir(signal_folder):
        if signal_file.startswith('signal') and signal_file.endswith('.csv'):
            label_file = signal_file.replace('signal', 'label')
            signal_path = os.path.join(signal_folder, signal_file)
            label_path = os.path.join(label_folder, label_file)
            
            if os.path.exists(label_path):
                processed_data.extend(process_ppg_file(signal_path, label_path, augment=augment))
            else:
                print(f"Warning: No matching label file for {signal_file}")
    
    df = pd.DataFrame(processed_data, columns=['PPG_Signal', 'Glucose_Level', 'Age', 'Gender'])
    df.to_pickle(output_file)
    print(f"Processed data saved to {output_file}. Dataset size: {len(df)} samples.")
```

```python
def plot_ppg_signals(pickle_file, num_samples=5):
    df = pd.read_pickle(pickle_file)
    plt.figure(figsize=(12, 6))
    for i in range(min(num_samples, len(df))):
        ppg_signal = df.iloc[i]["PPG_Signal"]
        plt.plot(ppg_signal, label=f"Sample {i+1} (Glucose: {df.iloc[i]['Glucose_Level']} mg/dL)")
    plt.xlabel("Time (samples at 30Hz)")
    plt.ylabel("Normalized PPG Amplitude")
    plt.title("Visualization of Processed PPG Signals")
    plt.legend()
    plt.show()
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, resample
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

def split_train_test(pickle_file, test_size=0.2, random_state=42):
    df = pd.read_pickle(pickle_file)
    df = df.sample(frac=1, random_state=random_state).reset_index(drop=True)  # Shuffle dataset
    
    X = np.array(df['PPG_Signal'].tolist())  # Convert list of arrays into a NumPy array
    y = df['Glucose_Level'].values  # Target variable
    
    scaler_X = MinMaxScaler()
    X = scaler_X.fit_transform(X)
    
    scaler_y = MinMaxScaler()
    y = scaler_y.fit_transform(y.reshape(-1, 1)).flatten()
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    print(f"Train set size: {len(X_train)}, Test set size: {len(X_test)}")
    return X_train, X_test, y_train, y_test, scaler_y
```

```python
# Cell 8: Train a Random Forest Model
def train_model(X_train, y_train, n_estimators=100, random_state=42):
    model = RandomForestRegressor(n_estimators=n_estimators, random_state=random_state)
    model.fit(X_train, y_train)
    return model
```

```python
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    print(f"Mean Absolute Error: {mae:.2f}")
    print(f"Root Mean Squared Error: {rmse:.2f}")
    return mae, rmse
```

```python
def build_optimized_lstm_model(input_shape):
    model = Sequential([
        Bidirectional(LSTM(256, return_sequences=True, activation='relu'), input_shape=input_shape),
        Bidirectional(LSTM(128, return_sequences=True)),
        Bidirectional(LSTM(64)),
        BatchNormalization(),
        Dropout(0.1),
        Dense(128, activation='relu'),
        Dense(64, activation='relu'),
        Dense(1)  # Output layer for regression
    ])
    model.compile(optimizer=Adam(learning_rate=0.0002), loss='mse', metrics=['mae'])
    return model
```

```python
def train_optimized_lstm(X_train, y_train, X_test, y_test, model_path='optimized_lstm_model2.h5', epochs=150, batch_size=32):
    input_shape = (X_train.shape[1], 1)
    X_train = np.expand_dims(X_train, axis=-1)  # Reshape for LSTM
    X_test = np.expand_dims(X_test, axis=-1)
    
    model = build_optimized_lstm_model(input_shape)
    
    early_stopping = EarlyStopping(monitor="val_loss", patience=15, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, min_lr=1e-6)
    
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=epochs, batch_size=batch_size, callbacks=[early_stopping, reduce_lr])
    model.save(model_path)
    print(f"Optimized model saved to {model_path}")
    return model
```

```python
def evaluate_model(model, X_test, y_test, scaler_y):
    X_test = np.expand_dims(X_test, axis=-1)  # Ensure correct shape for predictions
    y_pred = model.predict(X_test)
    
    # Inverse transform predictions to original glucose scale
    y_pred_original = scaler_y.inverse_transform(y_pred.reshape(-1, 1)).flatten()
    y_test_original = scaler_y.inverse_transform(y_test.reshape(-1, 1)).flatten()
    
    # Compute Metrics
    mae = mean_absolute_error(y_test_original, y_pred_original)
    rmse = np.sqrt(mean_squared_error(y_test_original, y_pred_original))
    r2 = r2_score(y_test_original, y_pred_original)
    
    print(f"Mean Absolute Error: {mae:.2f}")
    print(f"Root Mean Squared Error: {rmse:.2f}")
    print(f"R Score: {r2:.2f}")
    
    # Plot actual vs predicted values
    plt.figure(figsize=(8, 6))
    plt.scatter(y_test_original, y_pred_original, alpha=0.7, label='Predicted vs Actual')
    plt.plot([min(y_test_original), max(y_test_original)], [min(y_test_original), max(y_test_original)], 'r', linestyle='--')
    plt.xlabel("Actual Glucose Level")
    plt.ylabel("Predicted Glucose Level")
    plt.title("Predicted vs. Actual Glucose Levels")
    plt.legend()
    plt.show()
    
    return mae, rmse, r2
```

```python
def build_ensemble_lstm_model(input_shape):
    models = []
    for _ in range(3):  # Create 3 separate LSTM models
        model = Sequential([
            Bidirectional(LSTM(256, return_sequences=True, activation='relu'), input_shape=input_shape),
            Bidirectional(LSTM(128, return_sequences=True)),
            Bidirectional(LSTM(64)),
            BatchNormalization(),
            Dropout(0.1),
            Dense(128, activation='relu'),
            Dense(64, activation='relu'),
            Dense(1)  # Output layer for regression
        ])
        model.compile(optimizer=Adam(learning_rate=0.0002), loss='mse', metrics=['mae'])
        models.append(model)
    return models
```

```python
def train_ensemble_lstm(X_train, y_train, X_test, y_test, model_paths=['lstm_model_1.h5', 'lstm_model_2.h5', 'lstm_model_3.h5'], epochs=150, batch_size=32):
    input_shape = (X_train.shape[1], 1)
    X_train = np.expand_dims(X_train, axis=-1)  # Reshape for LSTM
    X_test = np.expand_dims(X_test, axis=-1)
    
    models = build_ensemble_lstm_model(input_shape)
    
    early_stopping = EarlyStopping(monitor="val_loss", patience=15, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, min_lr=1e-6)
    
    for i, model in enumerate(models):
        print(f"Training model {i+1}...")
        model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=epochs, batch_size=batch_size, callbacks=[early_stopping, reduce_lr])
        model.save(model_paths[i])
        print(f"Model {i+1} saved to {model_paths[i]}")
    
    return models
```

```python
def ensemble_predict(models, X_test):
    X_test = np.expand_dims(X_test, axis=-1)
    predictions = np.array([model.predict(X_test) for model in models])
    final_predictions = np.mean(predictions, axis=0)  # Average predictions from all models
    return final_predictions.flatten()
```

```python
import os
import pickle
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model

def save_ensemble_model(models, scaler_y, base_path='model_ensemble'):
    """
    Save an ensemble of models along with the scaler used for the target variable.
    
    Parameters:
    -----------
    models : list
        List of Keras models in the ensemble
    scaler_y : object
        The scaler used to transform the target variable
    base_path : str
        Base directory to save all model files
    """
    # Create the directory if it doesn't exist
    os.makedirs(base_path, exist_ok=True)
    
    # Save each model
    model_paths = []
    for i, model in enumerate(models):
        model_path = os.path.join(base_path, f'model_{i+1}.h5')
        model.save(model_path)
        model_paths.append(model_path)
    
    # Save the scaler
    scaler_path = os.path.join(base_path, 'scaler_y.pkl')
    with open(scaler_path, 'wb') as f:
        pickle.dump(scaler_y, f)
    
    # Save metadata about the ensemble
    metadata = {
        'num_models': len(models),
        'model_paths': model_paths,
        'scaler_path': scaler_path
    }
    
    metadata_path = os.path.join(base_path, 'metadata.pkl')
    with open(metadata_path, 'wb') as f:
        pickle.dump(metadata, f)
    
    print(f"Ensemble model saved to {base_path}")
    return metadata_path

def load_ensemble_model(metadata_path):
    """
    Load an ensemble of models along with the scaler used for the target variable.
    
    Parameters:
    -----------
    metadata_path : str
        Path to the metadata file
    
    Returns:
    --------
    models : list
        List of loaded Keras models
    scaler_y : object
        The scaler used to transform the target variable
    """
    # Load metadata
    with open(metadata_path, 'rb') as f:
        metadata = pickle.load(f)
    
    # Load models
    models = []
    for model_path in metadata['model_paths']:
        model = load_model(model_path)
        models.append(model)
    
    # Load scaler
    with open(metadata['scaler_path'], 'rb') as f:
        scaler_y = pickle.load(f)
    
    print(f"Loaded ensemble with {len(models)} models")
    return models, scaler_y
```

```python
def predict_glucose_levels(ppg_data, models, scaler_y):
    """
    Make predictions using the loaded ensemble model.
    
    Parameters:
    -----------
    ppg_data : numpy.ndarray
        Preprocessed PPG signal data
    models : list
        List of loaded Keras models
    scaler_y : object
        The scaler used to transform the target variable
    
    Returns:
    --------
    predictions : numpy.ndarray
        Predicted glucose levels
    """
    # Ensure data is properly shaped for LSTM
    if len(ppg_data.shape) == 2:
        ppg_data = np.expand_dims(ppg_data, axis=-1)
    
    # Make predictions with each model in the ensemble
    predictions = np.array([model.predict(ppg_data) for model in models])
    
    # Average predictions from all models
    final_predictions = np.mean(predictions, axis=0)
    
    # Inverse transform to get actual glucose levels
    glucose_levels = scaler_y.inverse_transform(final_predictions)
    
    return glucose_levels.flatten()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pickle
from scipy.signal import butter, filtfilt
import pywt
from tensorflow.keras.models import load_model

# Function to extract PPG signal from video
def extract_ppg_from_video(video_path, start_time=0, duration=10):
    """
    Extract PPG signal from video file
    
    Parameters:
    -----------
    video_path : str
        Path to the video file
    start_time : float
        Starting time in seconds
    duration : float
        Duration to extract in seconds
        
    Returns:
    --------
    ppg_signal : numpy.ndarray
        Extracted PPG signal
    """
    cap = cv2.VideoCapture(video_path)
    red_channel_values = []
    
    frame_rate = cap.get(cv2.CAP_PROP_FPS)
    start_frame = int(start_time * frame_rate)
    end_frame = int((start_time + duration) * frame_rate)
    
    cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)
    
    while cap.isOpened():
        frame_idx = int(cap.get(cv2.CAP_PROP_POS_FRAMES))
        if frame_idx >= end_frame:
            break
        ret, frame = cap.read()
        if not ret:
            break
        red_channel = frame[:, :, 2]
        avg_intensity = np.mean(red_channel)
        red_channel_values.append(avg_intensity)
    
    cap.release()
    return np.array(red_channel_values)

# Preprocessing functions
def bandpass_filter(signal, lowcut=0.5, highcut=8.0, fs=30, order=4):
    """Apply bandpass filter to signal"""
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, signal)

def bandpass_filter_narrow(signal, lowcut=0.5, highcut=4.0, fs=30, order=3):
    """Apply narrower bandpass filter to signal"""
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, signal)

def wavelet_denoise(signal, wavelet='db4', level=2, threshold_factor=3):
    """Apply wavelet denoising to signal"""
    coeffs = pywt.wavedec(signal, wavelet, level=level)
    coeffs[1:] = [pywt.threshold(c, np.std(c)/threshold_factor, mode='soft') for c in coeffs[1:]]  
    return pywt.waverec(coeffs, wavelet)

def adaptive_filter(signal, alpha=0.7):
    """Apply adaptive filtering to signal"""
    filtered_signal = np.zeros_like(signal)
    for i in range(1, len(signal)):
        filtered_signal[i] = alpha * filtered_signal[i-1] + (1-alpha) * signal[i]
    return filtered_signal

def normalize_signal(signal):
    """Normalize signal to zero mean and unit variance"""
    return (signal - np.mean(signal)) / np.std(signal)

def prepare_for_model(signal, expected_length=302):
    """Prepare signal for model input"""
    # Reshape for correct dimensions
    signal_reshaped = np.expand_dims(signal, axis=-1)
    
    # Trim or pad to expected length
    if signal_reshaped.shape[0] > expected_length:
        signal_reshaped = signal_reshaped[:expected_length]
    elif signal_reshaped.shape[0] < expected_length:
        signal_reshaped = np.pad(signal_reshaped, ((0, expected_length - signal_reshaped.shape[0]), (0, 0)), mode='constant')
    
    # Add batch dimension
    signal_reshaped = np.expand_dims(signal_reshaped, axis=0)
    return signal_reshaped

def load_ensemble_model(metadata_path):
    """Load ensemble model and scaler"""
    with open(metadata_path, 'rb') as f:
        metadata = pickle.load(f)
    
    models = []
    for model_path in metadata['model_paths']:
        model = load_model(model_path)
        models.append(model)
    
    with open(metadata['scaler_path'], 'rb') as f:
        scaler_y = pickle.load(f)
    
    return models, scaler_y

def ensemble_predict(models, X_test, scaler_y):
    """Make predictions with ensemble model"""
    # Remove extra dimension if present (we'll add it back)
    if len(X_test.shape) == 4:
        X_test = X_test[:, :, :, 0]
    
    # Make predictions
    predictions = np.array([model.predict(X_test) for model in models])
    final_predictions = np.mean(predictions, axis=0)
    
    # Convert to original scale
    original_predictions = scaler_y.inverse_transform(final_predictions)
    return original_predictions.flatten()[0]

# Main testing function
def test_preprocessing_variations(video_path, metadata_path='glucose_ensemble/metadata.pkl', known_glucose=None):
    """
    Test different preprocessing variations and return predictions
    
    Parameters:
    -----------
    video_path : str
        Path to the video file
    metadata_path : str
        Path to the model metadata file
    known_glucose : float (optional)
        Known glucose value for comparison
        
    Returns:
    --------
    results : dict
        Dictionary of preprocessing method and predicted glucose value
    """
    # Extract PPG signal
    ppg_signal = extract_ppg_from_video(video_path)
    
    # Plot raw signal
    plt.figure(figsize=(10, 4))
    plt.plot(ppg_signal)
    plt.title("Raw PPG Signal")
    plt.xlabel("Frame")
    plt.ylabel("Intensity")
    plt.show()
    
    # Load models
    models, scaler_y = load_ensemble_model(metadata_path)
    
    # Apply different preprocessing combinations
    variations = {}
    
    # 1. Bandpass only (wide)
    bp_wide = bandpass_filter(ppg_signal)
    variations["Bandpass (wide)"] = bp_wide
    
    # 2. Bandpass only (wide, inverted)
    bp_wide_inv = -bandpass_filter(ppg_signal)
    variations["Bandpass (wide, inverted)"] = bp_wide_inv
    
    # 3. Bandpass only (narrow)
    bp_narrow = bandpass_filter_narrow(ppg_signal)
    variations["Bandpass (narrow)"] = bp_narrow
    
    # 4. Bandpass only (narrow, inverted)
    bp_narrow_inv = -bandpass_filter_narrow(ppg_signal)
    variations["Bandpass (narrow, inverted)"] = bp_narrow_inv
    
    # 5. Bandpass + normalization
    bp_norm = normalize_signal(bandpass_filter(ppg_signal))
    variations["Bandpass + normalize"] = bp_norm
    
    # 6. Bandpass + normalization (inverted)
    bp_norm_inv = normalize_signal(-bandpass_filter(ppg_signal))
    variations["Bandpass + normalize (inverted)"] = bp_norm_inv
    
    # 7. Bandpass + wavelet
    bp_wavelet = wavelet_denoise(bandpass_filter(ppg_signal))
    variations["Bandpass + wavelet"] = bp_wavelet
    
    # 8. Bandpass + wavelet (inverted)
    bp_wavelet_inv = wavelet_denoise(-bandpass_filter(ppg_signal))
    variations["Bandpass + wavelet (inverted)"] = bp_wavelet_inv
    
    # 9. Full pipeline
    full = adaptive_filter(wavelet_denoise(bandpass_filter(ppg_signal)))
    variations["Full pipeline"] = full
    
    # 10. Full pipeline (inverted)
    full_inv = adaptive_filter(wavelet_denoise(-bandpass_filter(ppg_signal)))
    variations["Full pipeline (inverted)"] = full_inv
    
    # Plot all variations
    plt.figure(figsize=(14, 8))
    for i, (name, signal) in enumerate(variations.items()):
        plt.subplot(5, 2, i+1)
        plt.plot(signal[:300])  # Plot first 300 points
        plt.title(name)
        plt.tight_layout()
    plt.show()
    
    # Make predictions
    results = {}
    for name, signal in variations.items():
        # Scale signal (MinMaxScaler)
        scaler = MinMaxScaler()
        signal_scaled = scaler.fit_transform(signal.reshape(-1, 1)).flatten()
        
        # Prepare for model
        X = prepare_for_model(signal_scaled)
        
        # Predict
        pred = ensemble_predict(models, X, scaler_y)
        results[name] = pred
    
    # Display results
    result_df = pd.DataFrame(list(results.items()), columns=['Preprocessing', 'Prediction'])
    result_df['Prediction'] = result_df['Prediction'].round(1)
    
    # If known glucose provided, calculate errors
    if known_glucose is not None:
        result_df['Error'] = abs(result_df['Prediction'] - known_glucose).round(1)
        result_df = result_df.sort_values('Error')
    else:
        result_df = result_df.sort_values('Prediction')
    
    return result_df

# Function to analyze multiple videos with known glucose values
def analyze_multiple_videos(video_paths, known_values, metadata_path='glucose_ensemble/metadata.pkl'):
    """
    Analyze multiple videos with known glucose values
    
    Parameters:
    -----------
    video_paths : list
        List of paths to video files
    known_values : list
        List of known glucose values
    metadata_path : str
        Path to model metadata
        
    Returns:
    --------
    best_method : str
        Preprocessing method with lowest average error
    """
    results = {}
    for video_path, known_value in zip(video_paths, known_values):
        print(f"Processing {video_path} (Known value: {known_value} mg/dL)")
        df = test_preprocessing_variations(video_path, metadata_path, known_value)
        
        for _, row in df.iterrows():
            method = row['Preprocessing']
            error = row['Error']
            
            if method not in results:
                results[method] = []
            
            results[method].append(error)
    
    # Calculate average error for each method
    avg_errors = {}
    for method, errors in results.items():
        avg_errors[method] = np.mean(errors)
    
    # Sort methods by average error
    sorted_methods = sorted(avg_errors.items(), key=lambda x: x[1])
    
    # Display results
    result_df = pd.DataFrame(sorted_methods, columns=['Preprocessing', 'Avg Error'])
    
    return result_df
```

```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from tensorflow.keras.models import load_model
from scipy.signal import butter, filtfilt

def fix_ceg_with_direct_models(pickle_file, model_paths=['lstm_model_1.h5', 'lstm_model_2.h5', 'lstm_model_3.h5'], scaler_path='glucose_ensemble/scaler_y.pkl'):
    """
    Fix the Clarke Error Grid evaluation using direct model loading
    
    Parameters:
    -----------
    pickle_file : str
        Path to the training data pickle file
    model_paths : list
        List of paths to individual model files
    scaler_path : str
        Path to the output scaler
    """
    # Load training data
    print("Loading training data...")
    df = pd.read_pickle(pickle_file)
    
    # Extract PPG signals and glucose values
    ppg_signals = np.array(df['PPG_Signal'].tolist())
    glucose_values = df['Glucose_Level'].values
    
    # Load models directly
    print("Loading individual models...")
    models = []
    for path in model_paths:
        print(f"Loading model from {path}")
        model = load_model(path)
        models.append(model)
    
    # Load scaler
    print("Loading scaler...")
    with open(scaler_path, 'rb') as f:
        scaler_y = pickle.load(f)
    
    # Create a fresh MinMaxScaler for input data
    # This is CRUCIAL - we'll use the same approach as during training
    scaler_X = MinMaxScaler()
    
    # Make predictions
    print("Making predictions...")
    predictions = []
    
    for i, ppg in enumerate(ppg_signals):
        if i % 50 == 0:
            print(f"Processing sample {i}/{len(ppg_signals)}")
            
        # Preprocess the signal (apply your standard preprocessing)
        # Here using bandpass filter as an example
        preprocessed_signal = bandpass_filter(ppg)
        
        # Scale the entire signal
        # This mimics what you did during training
        # We use fit_transform for each signal individually
        scaled_signal = scaler_X.fit_transform(preprocessed_signal.reshape(-1, 1)).flatten()
        
        # Prepare for model
        input_data = np.expand_dims(scaled_signal, axis=(0, -1))
        
        # Make ensemble prediction
        model_predictions = []
        for model in models:
            pred = model.predict(input_data, verbose=0)
            model_predictions.append(pred)
        
        # Average predictions
        pred_scaled = np.mean(model_predictions, axis=0)
        
        # Convert back to original scale
        pred_original = scaler_y.inverse_transform(pred_scaled)[0][0]
        predictions.append(pred_original)
    
    predictions = np.array(predictions)
    
    # Calculate metrics
    mae = mean_absolute_error(glucose_values, predictions)
    rmse = np.sqrt(mean_squared_error(glucose_values, predictions))
    r2 = r2_score(glucose_values, predictions)
    
    print("\nResults with direct model loading approach:")
    print(f"MAE: {mae:.2f} mg/dL")
    print(f"RMSE: {rmse:.2f} mg/dL")
    print(f"R: {r2:.4f}")
    
    # Generate Clarke Error Grid
    zone_counts, percentages = clarke_error_grid(
        glucose_values, 
        predictions, 
        title_string="Clarke Error Grid - Direct Model Loading"
    )
    
    # Print zone percentages
    for zone, percentage in percentages.items():
        print(f"Zone {zone}: {percentage:.1f}%")
    
    # Plot predictions vs actual
    plt.figure(figsize=(10, 6))
    plt.scatter(glucose_values, predictions, alpha=0.7)
    plt.plot([min(glucose_values), max(glucose_values)], 
             [min(glucose_values), max(glucose_values)], 'r--')
    plt.xlabel("Actual Glucose Level (mg/dL)")
    plt.ylabel("Predicted Glucose Level (mg/dL)")
    plt.title(f"Predicted vs. Actual Glucose Levels\nR = {r2:.4f}")
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return {
        'MAE': mae,
        'RMSE': rmse,
        'R': r2,
        'CEG_zones': zone_counts,
        'CEG_percentages': percentages
    }

def bandpass_filter(signal, lowcut=0.5, highcut=8.0, fs=30, order=4):
    """Apply bandpass filter to signal"""
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, signal)

def clarke_error_grid(reference, prediction, title_string="Clarke Error Grid"):
    """
    Implement the Clarke Error Grid
    
    Parameters:
    -----------
    reference : array-like
        Reference glucose values (ground truth)
    prediction : array-like
        Predicted glucose values
    title_string : str
        Title for the plot
        
    Returns:
    --------
    zone_counts : dict
        Count of predictions in each zone (A, B, C, D, E)
    percentages : dict
        Percentage of predictions in each zone
    """
    # Ensure data is in numpy arrays
    reference = np.array(reference)
    prediction = np.array(prediction)
    
    # Create the figure and axes
    fig, ax = plt.subplots(1, 1, figsize=(10, 8))
    
    # Set up the plot
    max_glucose = max(reference.max(), prediction.max()) + 20
    min_glucose = min(reference.min(), prediction.min()) - 20
    
    ax.scatter(reference, prediction, c='blue', marker='o', alpha=0.8)
    ax.plot([0, max_glucose], [0, max_glucose], 'k--', label='Perfect prediction')
    
    # Zone A - Clinically accurate
    ax.fill_between([0, 70], [0, 70], [70, 180], color='green', alpha=0.1)
    ax.fill_between([70, 290], [70*0.8, 290*0.8], [70*1.2, 290*1.2], color='green', alpha=0.1)
    
    # Zone B - Benign errors
    ax.fill_between([70, 290], [0, 0], [70*0.8, 290*0.8], color='blue', alpha=0.1)
    ax.fill_between([70, 180], [70*1.2, 180*1.2], [400, 400], color='blue', alpha=0.1)
    ax.fill_between([180, 290], [180*1.2, 290*1.2], [400, 400], color='blue', alpha=0.1)
    
    # Zone C - Overcorrection errors
    ax.fill_between([130, 180], [0, 0], [130*0.8, 180*0.8], color='yellow', alpha=0.1)
    
    # Zone D - Dangerous failure to detect
    ax.fill_between([0, 70], [70, 180], [400, 400], color='red', alpha=0.1)
    
    # Zone E - Erroneous treatment errors
    ax.fill_between([70, 180], [0, 0], [70, 70], color='purple', alpha=0.1)
    
    # Labeling and formatting
    ax.set_xlim([min_glucose, max_glucose])
    ax.set_ylim([min_glucose, max_glucose])
    ax.set_xlabel("Reference Glucose (mg/dL)")
    ax.set_ylabel("Predicted Glucose (mg/dL)")
    ax.set_title(title_string)
    ax.grid(True)
    
    # Zone labels
    ax.text(30, 15, "A", fontsize=15)
    ax.text(30, 300, "D", fontsize=15)
    ax.text(90, 40, "E", fontsize=15)
    ax.text(200, 300, "B", fontsize=15)
    ax.text(150, 40, "C", fontsize=15)
    
    # Calculate points in each zone
    zone_a = 0
    zone_b = 0
    zone_c = 0
    zone_d = 0
    zone_e = 0
    
    for i in range(len(reference)):
        ref = reference[i]
        pred = prediction[i]
        
        # Zone A
        if (ref <= 70 and pred <= 70) or (ref >= 70 and pred >= 70 and pred <= ref*1.2 and pred >= ref*0.8):
            zone_a += 1
        # Zone B
        elif (ref >= 70 and ((pred <= ref*0.8 and pred >= 70) or (pred >= ref*1.2 and pred <= 400))) or \
             (ref <= 70 and pred <= 180 and pred >= 70) or \
             (ref >= 180 and pred <= 400 and pred >= ref*0.7):
            zone_b += 1
        # Zone C
        elif (ref >= 130 and ref <= 180 and pred <= 70) or \
             (ref > 180 and pred <= 70):
            zone_c += 1
        # Zone D
        elif ((ref >= 70 and ref <= 180) and (pred >= 180)) or \
             (ref <= 70 and pred >= 180):
            zone_d += 1
        # Zone E
        elif (ref >= 70 and pred <= 70) or \
             (ref <= 70 and pred >= 150):
            zone_e += 1
    
    zone_counts = {
        'A': zone_a,
        'B': zone_b,
        'C': zone_c,
        'D': zone_d,
        'E': zone_e
    }
    
    total = len(reference)
    percentages = {zone: count / total * 100 for zone, count in zone_counts.items()}
    
    # Add percentages to the plot
    stats_text = f"Zone A: {percentages['A']:.1f}%\n" \
                 f"Zone B: {percentages['B']:.1f}%\n" \
                 f"Zone C: {percentages['C']:.1f}%\n" \
                 f"Zone D: {percentages['D']:.1f}%\n" \
                 f"Zone E: {percentages['E']:.1f}%\n" \
                 f"Total: {len(reference)} points"
    
    ax.text(0.02, 0.98, stats_text, transform=ax.transAxes, 
            verticalalignment='top', horizontalalignment='left',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    plt.tight_layout()
    plt.show()
    
    return zone_counts, percentages

# Example usage
fix_ceg_with_direct_models(
    "processed_ppg_data.pkl", 
    model_paths=['lstm_model_1.h5', 'lstm_model_2.h5', 'lstm_model_3.h5'],
    scaler_path='glucose_ensemble/scaler_y.pkl'
)
```
